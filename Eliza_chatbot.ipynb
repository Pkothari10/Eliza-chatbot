```python
from nltk.tokenize import sent_tokenize,word_tokenize
filename2 = 'Harrypotter.txt'
file = open(filename2, 'rt')
text = file.read()
file.close()
```

```python
#1.2 (1 point) Tokenize sentences and words. Print lens of sentences and tokens and only print
#the first 20 words
```

```python
word_token =word_tokenize(text)
print("number of words:{}".format(len(word_token)))
print(word_token[:20])
sent_token =sent_tokenize(text)
print("number of sentences:{}".format(len(sent_token)))
```

```python
#1.3 (1 point) Remove punctuations. Only print the first 20 words
```

```python
tokens = [w.lower() for w in word_token]
import string
import re
words = re.split(r'\W+', text)
tokens = [w.lower() for w in words]
print(tokens[:20])
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
words1 = [word for word in stripped if word.isalpha()]
print("Number of words after punctuation:{}".format(len(words1)))

```

```python
#1.4 (1 point)Remove stop words and count the distinct cleaned words.
```

```python
from nltk.corpus import stopwords
from nltk.probability import FreqDist
stop_words = set(stopwords.words('english'))
Notstopwords= [w for w in words1 if not w in stop_words]
FreqDist1 =FreqDist(Notstopwords)
print("FreqDist with {} samples and {} outcomes".format(len(FreqDist1),len(Notstopwords)))
```

```python
#1.5 (1 point) Lemmatize the cleaned words and count the distinct lemmatized words. Whatâ€™s
#different from 1.4? Please clearly explain.
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_output =[lemmatizer.lemmatize(w) for w in Notstopwords]
Fdist =FreqDist(lemmatized_output)
print("FreqDist with {} samples and {} outcomes".format(len(Fdist),len(lemmatized_output)))
print(lemmatized_output[:20])
           
```
