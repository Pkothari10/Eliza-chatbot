```python
pronoun_dict = {
    "you": "me",
    "your": "my",
    "yours": "mine",
    "you've": "I have",
    "you'll": "I will",      
        }    
```

```python
def kill_detect(user_input):
    kill_reg =  r'.*(kill|murder|die).*'       
    ans = re.findall(kill_reg, user_input) 
```

```python
def transform_pronoun(match_word):    
    tokenizer = word_tokenize(match_word)    
    
    for index, token in enumerate(tokenizer):           
        if token in pronoun_dict:                       
            tokenizer[index] = pronoun_dict[token]      
    return ' '.join(tokenizer)                          
```

```python
def analyze_input(user_input):   
    
    for pattern, responses in reply_dict.items():   
        
        
        ans = re.search(pattern, user_input.rstrip(",.!"))   
        
        if ans:    
            response = random.choice(responses)
            
            
            if "%s" in response:                     
                match_word = ans[1]                  
                
            
                trans_response = transform_pronoun(match_word)  
                response = response %(trans_response) 
                return response
            else: 
                return response
```

```python
def input_answer(client):
        name = input('[%s] ' %client) 
        return name        
```

```python
def convo_body(client):
    print("[Eliza] Hi %s. How can I help you today?" %client, end = '\n')
        
    quits = [r'.*bye.*', r'.*quit.*', r'.*exit.*']                           
    
    while True:
        user_input = input_answer(client) 
        print()
        user_input = user_input.lower()   
        
        find_quits = 0
        
        for i in quits:
            matches = re.findall(i, user_input)
            find_quits += len(matches)            
             
        if find_quits == 0:                    
            kill_detect(user_input)         
            response = analyze_input(user_input)
            print("[Eliza] " + response, end = '\n')            
        else: 
            print("[Eliza] Bye. Have a wonderful day!") 
            break
```

```python
def initialize_chat():       
    print("[Eliza] Hi. I am a psychotherapist. What is your name?", end = '\n')

    client = "Name"                               
    name_input = input('[Name] ')
    print()
    
    name_input = re.sub(r'[^\w]', ' ', name_input)   

    tokenizer = word_tokenize(name_input)            

    
    if "name" not in name_input:           
        user_name = name_input             
    else:
        user_name = tokenizer[-1]          

    return user_name
    
        
# main function
def main():
   client = initialize_chat()       
   convo_body(client)             
    
if __name__ == "__main__":
    main()
```
